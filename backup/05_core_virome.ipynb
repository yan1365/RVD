```python
beef_final1 = beef_core.copy()
dairy_final1 = dairy_core.copy()
grazing_final1 = grazing_core.copy()
non_final1 = non_core.copy()
cattle_final1 = cattle_core.copy()
sheep_final1 = sheep_core.copy()
goat_final1 = goat_core.copy()
wild_final1 = wild_core.copy()
```

```python
## species
core_df_species = [cattle_core, sheep_core, goat_core, wild_core]

core_votu_species = []
for f in core_df_species:
    for i in f['votu']:
        core_votu_species.append(i) 
for f in [cattle_final1, sheep_final1, goat_final1, wild_final1]:
    f.set_index('votu', inplace = True)
    votu = list(f.index)
    for i in core_votu_species:
        if i not in votu:
            f.loc[i, 'per'] = 0
            
data_frames = [cattle_final1, sheep_final1, goat_final1, wild_final1]
core_virome_50_species = reduce(lambda  left,right: pd.merge(left,right,left_index=True, right_index=True), data_frames)
core_virome_50_species.columns = ['cattle', 'sheep', 'goat', 'wild']
```

```python
## beef dairy
core_df_beefdairy = [beef_core, dairy_core]

core_votu_beefdairy = []
for f in core_df_beefdairy:
    for i in f['votu']:
        core_votu_beefdairy.append(i) 
for f in [beef_final1, dairy_final1]:
    f.set_index('votu', inplace = True)
    votu = list(f.index)
    for i in core_votu_beefdairy:
        if i not in votu:
            f.loc[i, 'per'] = 0

core_virome_50_beefdairy = pd.merge(beef_final1, dairy_final1, left_index=True, right_index=True)
core_virome_50_beefdairy.columns = ['beef', 'dairy']
```

```python
## grazing
core_df_grazing = [grazing_core, non_core]

core_votu_grazing = []
for f in core_df_grazing:
    for i in f['votu']:
        core_votu_grazing.append(i) 
for f in [grazing_final1, non_final1]:
    f.set_index('votu', inplace = True)
    votu = list(f.index)
    for i in core_votu_grazing:
        if i not in votu:
            f.loc[i, 'per'] = 0

core_virome_50_grazing = pd.merge(grazing_final1, non_final1, left_index=True, right_index=True)
core_virome_50_grazing.columns = ['grazing', 'non-grazing']
```
